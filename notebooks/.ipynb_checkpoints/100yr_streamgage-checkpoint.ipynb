```python
import dataretrieval
import hyswap
import datetime
import matplotlib.pyplot as plt
import importlib
```

```python

site_no = "06902000" #Grand River Near Sumner, MO
site_no = "06909000" # MO River at Boonville, MO
site_no = "06910450" # MO River at Jefferson City, MO
site_no = "06909000" # MO River at Booneville, MO

df, _ = dataretrieval.nwis.get_dv(site_no,
                                  parameterCd="00060",
                                  start="1776-01-01",
                                  # end="202-09-30"
                                 )
```

```python
window = '7-day'

df_formatted = hyswap.rasterhydrograph.format_data(
    df, 
    '00060_Mean', 
    window_width=window
)
```

# Now Try `hyswap` Percentiles

```python
# Define percentile thresholds to use
percentile_thresholds=[5, 10, 25, 50, 75, 90, 95]

# calculate fixed percentiles for entire dataset
fixed_percentile_values = hyswap.percentiles.calculate_fixed_percentile_thresholds(
    df['00060_Mean'], 
    percentile_thresholds, 
    # window_width=window
)
```

```python
# calculate variable percentiles for each day of the year
variable_percentile_values = hyswap.percentiles.calculate_variable_percentile_thresholds_by_day(
    df, 
    "00060_Mean", 
    percentile_thresholds, 
    window_width=window
)
```

```python
# get year/doy information
df_year = hyswap.utils.define_year_doy_columns(df,
                                               year_type='water',
                                               clip_leap_day=True)

# plotting percentiles by day with line shade between
# fig, ax = plt.subplots(figsize=(10, 6))
fig, ax = plt.subplots(figsize=(10, 6))

# filter down to data from 2022
water_year = 2025
# df_water_year = df_year[df_year['index_year'] == water_year]
df_water_year = df_year[df_year['index_year'].isin(range(2019, 2026, 1))]

# plot data
ax = hyswap.plots.plot_duration_hydrograph(
    variable_percentile_values,
    df_water_year,
    "00060_Mean",
    ax=ax,
    # data_label=f"Water Year {water_year}",
    data_label=f"Measured Discharge",
    title=f"Percentiles of Streamflow by Day of Year - Site {site_no}"
)
plt.tight_layout()
# plt.show()

# plt.savefig(f'hyswap-percentiles-{site_no}-cy{water_year}.png')
plt.savefig(f'hyswap-percentiles-{site_no}.png')
```

```python
percentiles_predams = hyswap.percentiles.calculate_variable_percentile_thresholds_by_day(
    df[:'05-1937'], "00060_Mean", percentile_thresholds
)
```

```python
percentiles_postdams = hyswap.percentiles.calculate_variable_percentile_thresholds_by_day(
    df['1967':], "00060_Mean", percentile_thresholds
)
```

```python
## Convert to datetime. Add year with Feb 29. Going to ignore the year later. 
x = []
for i in percentiles_predams.index.values:
    i += '-2000' 
    x.append(datetime.strptime(i, "%m-%d-%Y"))
# datetime.strptime(date_string_3, "%m-%d")
```

```python
plt.figure(figsize=(10,6))
a = 0.75
# plt.plot(x, percentiles_predams.p25, c='C0', alpha=a)
# plt.plot(x, percentiles_predams.p50, c='C0', alpha=a)

plt.fill_between(x, 
                 percentiles_predams.p25.tolist(), 
                 percentiles_predams.p50.tolist(), 
                alpha=a, 
                label='pre  1938')

plt.fill_between(x, 
                 percentiles_postdams.p25.tolist(), 
                 percentiles_postdams.p50.tolist(), 
                alpha=a, 
                label='post 1968')

# plt.plot(x, percentiles_postdams.p25, c='C1',)
# plt.plot(x, percentiles_postdams.p50, c='C1')

plt.gca().xaxis.set_major_formatter(plt.matplotlib.dates.DateFormatter('%b'))
# plt.gca().xaxis.set_major_formatter(plt.matplotlib.dates.DateFormatter('%b-%Y'))

plt.legend()
plt.ylabel('Discharge [cfs]')
plt.title(f'IQR Comparison - {site_no} Missouri River at Boonville, MO')
# plt.savefig(f'100yr-IQR-reservoir-comparison-{site_no}.png')
```
