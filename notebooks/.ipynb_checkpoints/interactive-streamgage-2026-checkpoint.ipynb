````python
import os
import geopandas as gpd
import folium
import matplotlib.pyplot as plt
import dataretrieval

os.environ["API_USGS_PAT"] = "fVre04DhiUbQP3Zyqe04b13dPprvBvOgngdtfwhW"
## Had to add bc of .to_crs resulting in Inf issue
import pyproj
pyproj.network.set_network_enabled(False)
````

````python
crs = 4326

mo = gpd.read_file(
        r'C:\\Users\\nrthrac\\Desktop\\working\\MWSS\\python-gis\\gis-data\\MO_2014_County_Boundaries_shp'
        ).to_crs(crs)

path = r'P:\WRJC\PERSONAL\NRTHRAC\hydrologists-club\analysis\process-mo-usgs-gage-locations-2025'
fn = 'USGS-MoGages_2026_2025-12-16.csv'

df = gpd.pd.read_csv(
    os.path.join(path, fn), 
    dtype={'siteNumber': str},
    skipfooter=3,  
    engine='python', # c engine does not support skipfooter
)
df.siteNumber = df.siteNumber.str.strip()

iowa_gages = ['06903700', '06819185']
df = df[~df.siteNumber.isin(iowa_gages)]

df['sitepage_link'] = (r'<a href=https://waterdata.usgs.gov/monitoring-location/USGS-' + df.siteNumber + '>' + df.siteNumber + '</a>')
````

````python
## Used later to agrregate data
cols_to_add = ['fundingUSGSCMF', 'fundingUSGSAppropriated', 'fundingReimbursable']
cols_to_grp = list(df.columns[~df.columns.isin(cols_to_add)])

## Can group them and aggregate. Have to do something like this to keep all columns without adding lat/long. There may be a more ellegant solution. 
agg_add = {'fundingUSGSCMF': 'sum',
            'fundingUSGSAppropriated': 'sum',
            'fundingReimbursable': 'sum', 
            # 'customerName': 'sum',  # tried this and broke code
          }
agg_first = {col: 'first' for col in cols_to_grp} # get all other columns. This is useful when grouping specific gage. 
````

````python
## Convert USGS streamgage funding data into GeoDataFrame
## Aggregate to remove rows with blank geometry. .first() ignores NaNs so this should work fine for this application
## Added 'collectionCode' to grouping because there was at least one example ('06922500') of a gage having two collection codes for the same gage and customer name.  

df = df.groupby(['siteNumber', 'customerName', 'collectionCode']).agg(agg_add | agg_first)

gdf = gpd.GeoDataFrame(
    df, geometry=gpd.points_from_xy(df.longitude, df.latitude), crs=crs,
)
````

````python
# Various filters for data recieved from USGS email

filt_modnr = df.customerName.str.contains('Department of Natural Resources')
filt_wrc = df.customerName.str.contains('Water Resources Center')
# filt_notgw = df.collectionCategory != 'gw'
filt_sw = df.collectionCategory == 'sw'
# filt_cont = df.collectionCode.str.contains('CONT')
filt_cont = df.collectionCode.isin(['QCONT', 'STGCONT', 'ELEVCONT'])
````

````python
print(f'''
Unique MO Surface Water Gage count from USGS email: {len(df[filt_cont].siteNumber.unique())}
''')
````

````python
## Download State Site Data from dataretrieval. 
## UPDATED to use modern API
## bbox encapsulated Missouri, later refind down to only MO gages (for now)

df_api = dataretrieval.waterdata.get_time_series_metadata(bbox=[-96.042480,35.969115,-88.978271,40.672306], 
                                                        parameter_code=['00065', '00060', '62615'], 
                                                        # end='2025-01-01/..',
                                                        end='2025-01-01/..',
                                                        # last_modified='2025-01-01/..',
                                                       )[0]

````

````python
# df_api.columns
````

````python
df_api_filtered = df_api[
    (df_api.state_name == 'Missouri')
    # & (df_api.computation_period_identifier == 'Daily') 
    # & (df_api.computation_identifier == 'Instantaneous')
    ]

# ## Manually Remove 1 site, St Francis River at Fisk. See USGS email for more info. 
# email: "Re: [WMA ETO Jira] (WDFEEDBACK-809) [EXTERNAL] Python dataretrieval potential erroneous gages"
df_api_filtered = df_api_filtered[df_api_filtered.monitoring_location_id != 'USGS-07040000'] 
df_api_filtered

## shifted back to `df_site_info` because the following code was already formatted for it. Some refactoring to use the _api format would be helpful. 
df_site_info = df_api_filtered[['monitoring_location_id', 'geometry']].drop_duplicates()
df_site_info.monitoring_location_id = df_site_info.monitoring_location_id.str.split('-').str[1]
df_site_info.rename(columns = {'monitoring_location_id': 'siteNumber'}, inplace=True);

len(df_site_info)
````

````python
## This is a solution to get site information and put it into df df_site_info
## I updated the original method to be more Pythonic and not iterate over dataframe
## This is still an inelegant way to handle multiple calls to the API bc of size limitations. Eventually want to find a more proper way to do that.

test1 = dataretrieval.waterdata.get_monitoring_locations(monitoring_location_id = list(df_api_filtered.monitoring_location_id.drop_duplicates()[:200]))[0]
test2 = dataretrieval.waterdata.get_monitoring_locations(monitoring_location_id = list(df_api_filtered.monitoring_location_id.drop_duplicates()[200:]))[0]

## This is a more pythonic/pandas way of doing this than the looping I did originally
sort_col = 'monitoring_location_id'
df_location_info = gpd.pd.concat([test1, test2], ignore_index=True).sort_values(by=sort_col).reset_index(drop=True)

df_site_info = df_site_info.sort_values(by='siteNumber').reset_index(drop=True)
df_site_info['siteName'] = df_location_info.monitoring_location_name
df_site_info['site_tp_cd'] = df_location_info.site_type_code


## Looping method
# for i, row in test1.merge(test2, how='outer').iterrows():
#     filt = df_site_info.siteNumber == row.monitoring_location_id.split('-')[-1]
#     df_site_info.loc[filt, 'siteName'] = row.monitoring_location_name
#     df_site_info.loc[filt, 'site_tp_cd'] = row.site_type
````

## Plot interactive folium Map

````python
def get_tooltip_popup():
    tooltip = folium.GeoJsonTooltip(
        fields=['siteName', 'siteNumber'],)

    popup = folium.GeoJsonPopup(
        fields=['sitepage_link'])

    return tooltip, popup
````

Potentially another way to group without converting to GeoJson:  
`folium.map.FeatureGroup`

````python
## Plot with Interactive Folium Map

m = folium.Map(
                location=(38.36768, -92.47729), 
                zoom_start=6, 
                tiles="Cartodb Positron",
                # tiles="Stamen Watercolor",
                # tiles="CartoDB Voyager",
                # tiles="Cartodb dark_matter",
                # tiles=None,
)

folium.GeoJson(
              mo.boundary.to_json(), 
              name='MO Counties',
              style_function=lambda x: {
                  'weight': 1.5, 
                  'color': 'black', 
                  'alpha': 0.5}, 
              ).add_to(m)


tooltip, popup = get_tooltip_popup()

folium.GeoJson(
              gdf[filt_wrc & filt_sw].to_json(), 
              name='wrc & sw',
              tooltip=tooltip,
              popup=popup,
              marker = folium.Circle(
                  color='black'
              )
         ).add_to(m)


for site_tp, color in zip(['ST','SP','LK'],
                          ['blue', 'green', 'orange']):
    tooltip, popup = get_tooltip_popup()
    folium.GeoJson(
        df_site_info[df_site_info.site_tp_cd == site_tp].to_json(),
        # df_what[df_what.site_tp_cd == site_tp].to_json(),
        name=f'USGS API {site_tp}',
        tooltip=tooltip,
        popup=popup,
        marker = folium.Circle(
            radius=500,
            color=color,
            # color='green'
            # style_function = style_function,
        )
    ).add_to(m)

tooltip, popup = get_tooltip_popup()

folium.GeoJson(
    # gdf[filt_sw & filt_cont][filt_duplicates].to_json(),
    # gdf[filt_cont].drop_duplicates().to_json(),
    gdf[filt_cont].groupby('siteName', as_index=False).first().to_json(),
    name='USGS Email - filt_cont',
    tooltip=tooltip,
    popup=popup,
    marker = folium.Circle(
        color='red',
        radius=1000,
        fill_opacity=0.5, 
    )
).add_to(m)

folium.LayerControl().add_to(m)

# m.save('test-interactive-streamgage.html')
m
````
